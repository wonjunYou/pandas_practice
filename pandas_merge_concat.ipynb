```python
import pandas as pd

df= pd.DataFrame({
    "id": [1,2,3],
    "customer_id": [1,2,3],
    "customer_name": ['Robert','Peter','Dave']
})

df2= pd.DataFrame({
    "id": [1,2,3],
    "order_id": [100,200,300],
    "order_date": ['2021-01-21','2021-02-03','2020-10-01']
})
df
```

```python
#DATA FRAME 합치기 1) concat, 2) merge
#concat: 두 데이터 프레임을 위/아래 혹은 왼/오른쪽으로 연결하기만 한다.
#pd.concat([df1,df2])
pd.concat([df,df2])
```

```python
doc = pd.concat([df,df2], axis=1)
doc.head()
```

```python
pd.merge(df,df2)
```

```python
pd.merge(df,df2,on='id') #merge 기준 컬럼 명시
```

```python
#내부 JOIN (Inner join)
#on의 컬럼값이 두 데이터프레임에서 동일한 행 찾기✔
# 각 동일한 행의 컬럼/ 컬럼값만 가져오기.✔

#외부 JOIN(outer join)
#모든 데이터를 가져온다. 비는 부분은 결측치 NaN을 넣는다.
```

```python
#Left JoIN
#첫 번째 데이터프레임을 모두 가져오고, 왼쪽 기준으로 동일한 것은 붙이고, 아닌 것은 nan 처리.
#Right JOIN은 
#두 번째 데이터프레임을 모두 가져오고, ..

df = df.set_index("id")
df
```

```python
df2 = df2.set_index('id')
df2
```

```python
pd.merge(df,df2,left_index=True, right_index=True)
```

```python
pd.merge(df,df2,how="outer", left_index=True, right_index=True)
```
